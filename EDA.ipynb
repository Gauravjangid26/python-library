```python
# Bivariate Analysis 
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#1. scatter plot (numerical column,numerical column)
sns.scatterplot(x=tips['total_bill'],y=tips["tip"])
```

```python
#he attribute that distinguishes different categories within a dataset
sns.scatterplot(x=tips['total_bill'],y=tips["tip"],hue=tips['sex'],style=tips['smoker'])
```

```python
# heatmap(categorical,catgorical)
#p=pd.crosstab(column1,column2) work as pivot table
p=pd.crosstab(tips['day'],tips['time'])
p
```

```python
sns.heatmap(p)
```

```python
#groupby=aggregation of same categories
#df['Pclass'].mean() if we want mean of every categories
df.groupby('Pclass').mean()
```

```python
df.info()
```

```python
df1=df.drop(columns=['Name','Sex','Ticket','Cabin','Embarked'])
df1=df1.dropna()
```

```python
(df1.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
(df1.groupby('Pclass').mean()['Survived']*100).plot(kind='pie',autopct='%0.2f')
```
